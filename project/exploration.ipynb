```python
import sqlite3
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def read_table(table_name):   
    
    con = sqlite3.connect("../data/data.sqlite")

    query = f"SELECT * FROM {table_name}"
    df = pd.read_sql_query(query, con)
    
    con.close()  
    
    
    return df
```

```python
df_pt_t1=read_table('public_transport_type_1')
df_pt_t2=read_table('public_transport_type_2')
df_pt_t3=read_table('public_transport_type_3')
df_td=read_table('traffic_density')
df_wth=read_table('weather')
```

```python
df_concatenated = pd.concat([df_pt_t1, df_pt_t2, df_pt_t3])

df_pivot = df_concatenated.pivot_table(index='DATE_TIME', columns='transport_type_id', values='number_of_passage', fill_value=0)

df_pivot.columns = [f'type_{col}' for col in df_pivot.columns]

df_pivot['total_number_of_passage'] = df_pivot.sum(axis=1)

df_pivot.reset_index(inplace=True)

df_combined = pd.merge(df_pivot, df_td, on="DATE_TIME", how="inner")
df_wth["time"] = pd.to_datetime(df_wth["time"])
df_combined["DATE_TIME"] = pd.to_datetime(df_combined["DATE_TIME"])

df_final = pd.merge(df_combined, df_wth, left_on="DATE_TIME", right_on="time", how="inner")


df_daily = df_final.groupby(df_final['DATE_TIME'].dt.date).agg({
    'type_1': 'sum',
    'type_2': 'sum',
    'type_3': 'sum',
    'total_number_of_passage': 'sum',
    'NUMBER_OF_VEHICLES': 'sum',
    'AVERAGE_SPEED': 'mean',
    'temperature_2m (°C)': 'mean',
    'relativehumidity_2m (%)': 'mean',
    'dewpoint_2m (°C)': 'mean',
    'apparent_temperature (°C)': 'mean',
    'precipitation (mm)': 'mean',
    'rain (mm)': 'mean',
    'snowfall (cm)': 'mean',
    'snow_depth (m)': 'mean',
    'windspeed_10m (km/h)': 'mean',
    'windspeed_100m (km/h)': 'mean'
}).reset_index()


df_daily['passage_change_percentage'] = (df_daily['total_number_of_passage'] / df_daily['total_number_of_passage'].mean() - 1) * 100
df_daily['traffic_change_percentage'] = (df_daily['NUMBER_OF_VEHICLES'] / df_daily['NUMBER_OF_VEHICLES'].mean() - 1) * 100

```

```python
df_daily.head()
```

```python
variables = ['temperature_2m (°C)', 'windspeed_10m (km/h)', 'rain (mm)', 'snowfall (cm)']
titles = ['temperature', 'wind', 'rain', 'snow']
change_vars = ['traffic_change_percentage', 'passage_change_percentage']

for change_var, title in zip(variables, titles):
    # Plotting the graph
    fig, ax1 = plt.subplots(figsize=(12, 6))

    # Public Transportation Usage
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Public Transportation Usage', color='tab:red')
    ax1.plot(df_daily.index, df_daily['passage_change_percentage'], color='tab:red')
    ax1.tick_params(axis='y', labelcolor='tab:red')

    # Traffic Density
    ax2 = ax1.twinx()
    ax2.set_ylabel('Traffic Density', color='tab:blue')
    ax2.plot(df_daily.index, df_daily['traffic_change_percentage'], color='tab:blue')
    ax2.tick_params(axis='y', labelcolor='tab:blue')

    # Second axis
    ax3 = ax1.twinx()
    ax3.spines['right'].set_position(('outward', 60))
    ax3.set_ylabel('Variable: {}'.format(change_var), color='tab:green')
    ax3.plot(df_daily.index, df_daily[change_var], color='green')
    ax3.tick_params(axis='y', labelcolor='tab:green')

    fig.tight_layout()
    plt.title('{} - {}'.format(title, change_var.capitalize()))
    plt.show()
```

```python
variables = ['temperature_2m (°C)', 'windspeed_10m (km/h)', 'rain (mm)', 'snowfall (cm)']
titles = ['Temperature', 'Wind Speed', 'Rainfall', 'Snowfall']
colors = ['tab:blue'] * len(variables)
change_vars = ['NUMBER_OF_VEHICLES', 'total_number_of_passage']

for change_var in change_vars:
    for var, title, color in zip(variables, titles, colors):
        fig, ax = plt.subplots(figsize=(10, 6))

        ax.scatter(df_daily[change_var], df_daily[var], color=color)

        ax.set_xlabel(change_var)
        ax.set_ylabel('Variable: {}'.format(var))
        ax.set_title('{} - {}'.format(title, change_var))

        plt.show()
```

```python
variables = ['temperature_2m (°C)', 'windspeed_10m (km/h)', 'rain (mm)', 'snowfall (cm)']
titles = ['Air Temperature', 'Wind Speed', 'Rainfall Amount', 'Snowfall']

for var, title in zip(variables, titles):
    correlation_matrix = df_daily[[var, 'NUMBER_OF_VEHICLES', 'type_1', 'type_2', 'type_3', 'total_number_of_passage']].corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
    plt.title('Correlation Matrix: {}'.format(title))
    plt.show()
```

```python
correlation_matrix = df_daily[['temperature_2m (°C)', 'windspeed_10m (km/h)', 'rain (mm)', 'snowfall (cm)', 'NUMBER_OF_VEHICLES']].corr()

target_feature = 'NUMBER_OF_VEHICLES'  
correlations_with_target = correlation_matrix[target_feature].drop(target_feature)

print(f'Correlations with "{target_feature}" under specific conditions:')
print(correlations_with_target)
```

```python
correlation_matrix = df_daily[['temperature_2m (°C)', 'windspeed_10m (km/h)', 'rain (mm)', 'snowfall (cm)', 'total_number_of_passage']].corr()

target_feature = 'total_number_of_passage' 
correlations_with_target = correlation_matrix[target_feature].drop(target_feature)

print(f'Correlations with "{target_feature}" under specific conditions:')
print(correlations_with_target)
```

```python
def plot_scatter(x, y, hue, size, data, title):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=x, y=y, hue=hue, size=size, data=data, palette='coolwarm', sizes=(50, 200))
    plt.title(title)
    plt.show()

plot_scatter('rain (mm)', 'temperature_2m (°C)', 'NUMBER_OF_VEHICLES', 'NUMBER_OF_VEHICLES', df_daily, 'Relationship Between Rainfall and Temperature with Traffic Density')

plot_scatter('windspeed_10m (km/h)', 'temperature_2m (°C)', 'NUMBER_OF_VEHICLES', 'NUMBER_OF_VEHICLES', df_daily, 'Relationship Between Wind Speed and Temperature with Traffic Density')

plot_scatter('windspeed_10m (km/h)', 'rain (mm)', 'NUMBER_OF_VEHICLES', 'NUMBER_OF_VEHICLES', df_daily, 'Relationship Between Wind Speed and Rainfall with Traffic Density')

plot_scatter('snowfall (cm)', 'temperature_2m (°C)', 'NUMBER_OF_VEHICLES', 'NUMBER_OF_VEHICLES', df_daily, 'Relationship Between Snowfall and Temperature with Traffic Density')

plot_scatter('rain (mm)', 'temperature_2m (°C)', 'total_number_of_passage', 'total_number_of_passage', df_daily, 'Relationship Between Rainfall and Temperature with Public Transportation Usage')

plot_scatter('windspeed_10m (km/h)', 'temperature_2m (°C)', 'total_number_of_passage', 'total_number_of_passage', df_daily, 'Relationship Between Wind Speed and Temperature with Public Transportation Usage')

plot_scatter('windspeed_10m (km/h)', 'rain (mm)', 'total_number_of_passage', 'total_number_of_passage', df_daily, 'Relationship Between Wind Speed and Rainfall with Public Transportation Usage')

plot_scatter('snowfall (cm)', 'temperature_2m (°C)', 'total_number_of_passage', 'total_number_of_passage', df_daily, 'Relationship Between Snowfall and Temperature with Public Transportation Usage')
```
